```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_sc = sc.fit_transform(x)
```

```python
x_sc
```

```python
from sklearn.linear_model import Lasso
```

```python
names = x.columns
```

```python
names
```

```python
def lasso(alpha):
    df1 = pd.DataFrame()  # Fix the typo here
    df1["FeatureName"] = names
    for alpha in alpha:
        lasso = Lasso(alpha=alpha)
        lasso.fit(x_sc, y)
        column_names = 'Alpha = %f' % alpha
        df1[column_names] = lasso.coef_
    return df1

lasso([0.0001, 0.001, 0.1, 0.5, 1, 10, 100])
```
