## KNeighborsClassifier

```python
# from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression , LogisticRegression
```

```python
dataset = load_iris()
```

```python
X = dataset.data
# X
y = dataset.target
y
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(xtrain, ytrain)
```

```python
predictions = knn.predict(xtest)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(knn,xtrain,ytrain,scoring='accuracy')
print("Average Accuracy:", scores.mean())
print("Standard Deviation:", scores.std())
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, predictions)
```

### using grid search for KNeighborsClassifier

```python
# Find best parameters of Kneighbors using gridSearchCV
from sklearn.model_selection import GridSearchCV

params = [{'weights': ['uniform','distance'], 'n_neighbors':[3,4,5]}]
grid_search = GridSearchCV(knn,params,cv=5,scoring='accuracy',return_train_score=True,verbose=3)
```

```python
grid_search.fit(xtrain,ytrain)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
final_knn_model = grid_search.best_estimator_
final_score = cross_val_score(final_knn_model,xtrain,ytrain,scoring='accuracy')
print("Average Accuracy:", final_score.mean())
print("Standard Deviation:", final_score.std())
```

```python
final_knn_model = grid_search.best_estimator_
final_knn_model.fit(xtrain,ytrain)
```

```python
final_predictions = final_knn_model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, final_predictions)
```

## KNeighborsRegressor

```python
import pandas as pd
```

```python
data = pd.read_csv("BostonHousing.csv")
```

```python
data.head(3)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler(feature_range=(0, 1))
```

```python
x = data.iloc[:,:-1]
```

```python
x = sc.fit_transform(x)
x
```

```python
x.shape
```

```python
y = data.iloc[:,-1]
y.shape
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn_reg = KNeighborsRegressor()
knn_reg.fit(xtrain, ytrain)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(knn_reg, xtrain, ytrain, cv=5, scoring='neg_mean_squared_error')
print("Average Accuracy:", -scores.mean())
print("Standard Deviation:", scores.std())
```

```python
predictions = knn_reg.predict(xtest)
```

```python
from sklearn.metrics import r2_score
r2_score(ytest, predictions)
```

## using grid search for KNeighborsRegressor

```python
# Find best parameters of kneighbors using gridSearchCV
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_neighbors': [3,5, 7, 9],
    'weights': ['uniform', 'distance'],
}

grid_search = GridSearchCV(knn_reg,param_grid,scoring='neg_mean_squared_error',cv=5)
grid_search.fit(xtrain,ytrain)
```

```python
grid_search.best_params_ #{'n_neighbors': 3, 'weights': 'distance'}
-grid_search.best_score_ # 21.59207643565447
```

```python
final_knn_reg_model = grid_search.best_estimator_
scores = cross_val_score(final_knn_reg_model, xtrain, ytrain, cv=5, scoring='neg_mean_squared_error')
print("Average Accuracy:", -scores.mean())
print("Standard Deviation:", scores.std())
```

```python
final_knn_reg_model.fit(xtrain,ytrain)
```

```python
final_prediction = final_knn_reg_model.predict(xtest)
final_prediction
```

```python
r2_score(ytest,final_prediction)
```
