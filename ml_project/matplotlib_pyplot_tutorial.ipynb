```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4],[1,6,10,18],'ro')
plt.xlabel('hello')
plt.ylabel('world')
plt.axis((0,5,0,20))
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
