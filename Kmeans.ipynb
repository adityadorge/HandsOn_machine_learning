```r
library(ggplot2)
library(cluster)
```

```r
df = as.data.frame(read.csv("marksheet.csv"))
```

```r
df
```

```r
kmdata = df[,2:4]
kmdata
```

```r
wss = numeric(8)
for (k in 1:8) wss[k] = sum(kmeans(kmdata,centers = k,nstart = 25)$withinss)
plot(1:8,wss,type="b",xlab="number of cluster",ylab="wss value")
final_model = kmeans(kmdata,3,nstart=25)
final_model
```
